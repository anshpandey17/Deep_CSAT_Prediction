## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Clean Column Names:
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
df.head(1)
````

````python
# Dropping columns with extremely high missing values or no predictive importance

drop_columns = [
    'unique_id',
    'order_id',
    'connected_handling_time',
    'customer_city',
    'product_category',
    'item_price',
    'order_date_time'
]

df = df.drop(columns=drop_columns)

df.head()
````

````python
# Handling missing values in Customer Remarks

df['customer_remarks'] = df['customer_remarks'].fillna("No Comment")

df.isnull().sum()
````

````python
# Converting date columns to datetime format

df['issue_reported_at'] = pd.to_datetime(df['issue_reported_at'], dayfirst=True)
df['issue_responded'] = pd.to_datetime(df['issue_responded'], dayfirst=True)
df['survey_response_date'] = pd.to_datetime(df['survey_response_date'], dayfirst=True)
df.dtypes
````

````python
# Creating response time feature

df['response_time'] = (
    df['issue_responded'] - df['issue_reported_at']
).dt.total_seconds()

df.head()
````

````python
# Creating survey delay feature

df['survey_delay'] = (
    df['survey_response_date'] - df['issue_responded']
).dt.total_seconds()
````

````python
# Extract hour of issue reporting
df['issue_hour'] = df['issue_reported_at'].dt.hour

# Extract day of week
df['issue_day'] = df['issue_reported_at'].dt.dayofweek

# Weekend indicator
df['is_weekend'] = df['issue_day'].apply(lambda x: 1 if x >= 5 else 0)
````

````python
# Dropping original timestamp columns after feature extraction

df = df.drop(columns=[
    'issue_reported_at',
    'issue_responded',
    'survey_response_date'
])

df.head()
````

## ***5. Hypothesis Testing***

### Hypothetical Statement 1 - Does Response Time Affect Customer Satisfaction?

H0 (Null Hypothesis): Response time has no effect on customer satisfaction.

H1 (Alternative Hypothesis): Response time significantly affects customer satisfaction.

````python
from scipy.stats import f_oneway

groups = [group["response_time"].values for name, group in df.groupby("csat_score")]

f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("p-value:", p_value)
````

### Statistical Test Used and Justification

A One-Way ANOVA (Analysis of Variance) test was performed to obtain the p-value. This test compares the mean response times across multiple CSAT score groups to determine whether there is a statistically significant difference between them.

The One-Way ANOVA test was chosen because the analysis involves comparing the mean values of a numerical variable (`response_time`) across more than two independent groups defined by a categorical variable (`csat_score` with values 1–5). ANOVA is appropriate for determining whether there are statistically significant differences between these group means.

### Observation

The ANOVA test produced a p-value close to 0, which is significantly lower than the significance level of 0.05. Therefore, the null hypothesis is rejected.

This indicates that response time varies significantly across different CSAT score groups, suggesting that response time may influence customer satisfaction.

### Hypothesis Statement – 2

H0 (Null Hypothesis): Customer satisfaction (CSAT score) is independent of the agent shift.

H1 (Alternative Hypothesis): Customer satisfaction (CSAT score) is influenced by the agent shift.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import chi2_contingency

cont_table = pd.crosstab(df['agent_shift'], df['csat_score'])

chi2, p, dof, expected = chi2_contingency(cont_table)

print("Chi-square statistic:", chi2)
print("p-value:", p)
````

### Statistical Test Used and Justification

A Chi-Square Test of Independence was performed to obtain the p-value. This test examines whether there is a significant relationship between two categorical variables.

The Chi-Square test was chosen because both variables involved in the analysis (`agent_shift` and `csat_score`) are categorical. This test helps determine whether customer satisfaction is associated with the shift during which the issue was handled.

### Observation

The Chi-Square test produced a p-value of 4.95e-30, which is significantly smaller than the significance level of 0.05. Therefore, the null hypothesis is rejected.

This indicates that customer satisfaction is not independent of the agent shift. In other words, the shift during which customer issues are handled may influence the CSAT score.

### Hypothesis Statement – 3

H0 (Null Hypothesis): Customer satisfaction (CSAT score) is independent of the service channel.

H1 (Alternative Hypothesis): Customer satisfaction (CSAT score) is influenced by the service channel.

````python
# Perform Statistical Test to obtain P-Value
cont_table = pd.crosstab(df['channel_name'], df['csat_score'])

chi2, p, dof, expected = chi2_contingency(cont_table)

print("Chi-square statistic:", chi2)
print("p-value:", p)
````

### Statistical Test Used and Justification

A Chi-Square Test of Independence was used to obtain the p-value. This test evaluates whether there is a significant association between two categorical variables.

The test was selected because both `channel_name` and `csat_score` are categorical variables. The Chi-Square test helps determine whether the service channel through which a customer interacts with support influences their satisfaction rating.

### Observation

The Chi-Square test produced a p-value of 2.47e-36, which is far below the significance level of 0.05. Therefore, the null hypothesis is rejected.

This suggests that customer satisfaction is associated with the service channel used. Different communication channels may lead to varying levels of customer satisfaction.

# DEEP LAERNING MODEL :


 ## ANN - ARTIFICIAL NEAURAL NETWEORK
 ### Why ANN was used

Artificial Neural Networks (ANN) were used because they can learn complex nonlinear relationships between features. Since the dataset contains many engineered and text-derived features, ANN can capture hidden patterns that simpler models may miss.



### Model Architecture
##### Building the Model

A feedforward neural network was implemented using dense layers. The model includes two hidden layers (256 and 128 neurons) with ReLU activation to learn complex feature interactions. A softmax output layer was used for multi-class CSAT prediction.

### Regularization

Dropout layers were added between dense layers to reduce overfitting. Dropout randomly disables neurons during training, which helps the model generalize better on unseen data.

### Model Compilation

The model was compiled using the Adam optimizer and sparse categorical crossentropy loss. Adam provides efficient gradient updates, while the chosen loss function is suitable for multi-class classification problems.

````python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential([
    Dense(256, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.3),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(5, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
````

### Model Training

- The model was trained for 10 epochs with a batch size of 256. A validation split of 20% was used to monitor model performance during training and ensure that the model does not overfit the training data.
- We used sampled training data so it trains quickly.




````python
history = model.fit(
    X_sample,
    y_sample,
    epochs=10,
    batch_size=256,
    validation_split=0.2,
    verbose=1
)
````

### Training Performance

During training, the model gradually improved its accuracy across epochs. Validation accuracy stabilized around 42%, indicating that the model learned meaningful patterns while maintaining reasonable generalization.

### ANN Training Performance

The plot below shows the training and validation accuracy of the Artificial Neural Network (ANN) across epochs. It helps evaluate how well the model learns over time and whether overfitting occurs.

````python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("ANN Training Performance")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["Training Accuracy", "Validation Accuracy"])

plt.show()
````
